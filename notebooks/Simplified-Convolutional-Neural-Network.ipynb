```python
import rust_machine_learning
from syed_ml_lib import *
import os
```

```python
# Setup the CNN
num_classifications = 10
input_width = 28
input_height = 28
input_depth = 1
filters_per_conv_layer = [8]
filter_dimension = 3

CNN = rust_machine_learning.SimplifiedConvolutionalNeuralNetwork(num_classifications, input_width, input_height, input_depth, filters_per_conv_layer, filter_dimension)
```

```python
# Grab mnist data
import random
import numpy as np

df = pd.read_csv("../data/digit-recognizer/train.csv")

# Normalize images
df.iloc[:, 1:] = df.iloc[:, 1:].div(255.0)
images = df.iloc[:, 1:].values.tolist()
df.iloc[:, 0] = df.iloc[:, 0].astype(np.float32)
labels = df.iloc[:, 0].values.tolist()

df
```

```python
def test_mnist():
    num_batches_to_test = 500
    num_right = 0
    for i in range(num_batches_to_test):

        # Random sample one image from python list of images
        rand_index = random.randint(0, len(images) - 1)
        image = images[rand_index]
        label = labels[rand_index]

        prediction = CNN.classify([[image]])[0]
        if prediction == label:
            num_right += 1

    print("Accuracy: " + str(100.0 * num_right / (num_batches_to_test)))
```

```python
# Now train the model
learning_rate = 1e-4
batch_size = 1
num_iter = 3000
test_stride = 10

num_tests = num_iter // test_stride
train_stide = num_iter // num_tests

print("Accuracy before training:")
# test_cnn() # Should be around 10%
test_mnist()
print("Training...")
for i in range(num_tests):
    # CNN.train_using_image_loader(learning_rate, batch_size, train_stide)
    rand_index = random.randint(0, len(images) - 1)
    image = images[rand_index]
    label = labels[rand_index]

    # Train using MNIST dataset, select one image at a time randomly
    for i in range(train_stide):
        CNN.train_raw_data([[image]], [label], learning_rate)

    test_mnist()
```
